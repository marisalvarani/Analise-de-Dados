# Converter colunas

```python
#Converter coluna "in_deezer_playlists" para numerico
df['in_deezer_playlists'] = pd.to_numeric(df['in_deezer_playlists'], errors='coerce')
```

```python
#Verifica quantidade de valores distintos na coluna streams
print(df['streams'].nunique())
print(df['streams'].unique())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Selecionando as colunas numéricas do DataFrame
numeric_columns = df.select_dtypes(include=['number']).columns

# Definindo o tamanho da figura com base no número de variáveis numéricas
fig, axs = plt.subplots(nrows=(len(numeric_columns) // 3) + 1, ncols=3, figsize=(15, 5 * (len(numeric_columns) // 3 + 1)))

# Fazendo o loop para plotar cada variável numérica com boxplot
for i, col in enumerate(numeric_columns):
    ax = axs[i // 3, i % 3]  # Ajusta a posição do gráfico na grid
    ax.boxplot(df[col], vert=False, patch_artist=True, notch=True, boxprops=dict(facecolor='skyblue', color='black'))
    ax.set_title(f'{col} Boxplot')
    ax.set_xlabel(col)

# Ajustando o layout para evitar sobreposição
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Selecionando as variáveis categóricas específicas
categorical_columns = ['artist(s)_name', 'key', 'mode']

# Ajustando o tamanho da figura
fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(18, 6))

# Fazendo o loop para plotar gráficos de barras para cada uma das variáveis
for i, col in enumerate(categorical_columns):
    ax = axs[i]
    if col == 'artist(s)_name':
        # Para artist(s)_name, mostrar apenas os top 10 artistas
        df[col].value_counts().head(10).plot(kind='bar', ax=ax, color='skyblue', edgecolor='black')
    else:
        df[col].value_counts().plot(kind='bar', ax=ax, color='skyblue', edgecolor='black')

    ax.set_title(f'{col} Bar Plot')
    ax.set_xlabel(col)
    ax.set_ylabel('Count')
    ax.tick_params(axis='x', rotation=45)  # Rotaciona os rótulos do eixo X para melhorar a legibilidade

# Ajustando o layout para evitar sobreposição
plt.tight_layout()
plt.show()
```

# Análise Multivariada

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Ajustando o estilo do gráfico
sns.set(style="whitegrid")

# 1. Streams por released_date (utilizando released_date no formato datetime)
df['released_date'] = pd.to_datetime(df['released_date'])
plt.figure(figsize=(10, 6))
plt.plot(df['released_date'], df['streams'], marker='o', linestyle='-', color='b', alpha=0.7)
plt.title('Streams por Released Date', fontsize=16)
plt.xlabel('Released Date', fontsize=12)
plt.ylabel('Streams', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()

# 2. Streams por BPM (dispersão)
plt.figure(figsize=(10, 6))
plt.scatter(df['bpm'], df['streams'], color='green', alpha=0.5)
plt.title('Streams por BPM', fontsize=16)
plt.xlabel('BPM', fontsize=12)
plt.ylabel('Streams', fontsize=12)
plt.tight_layout()

# 3. Track Name por Streams (Top 10 músicas por Streams)
top_tracks = df[['track_name', 'streams']].sort_values(by='streams', ascending=False).head(10)
plt.figure(figsize=(10, 6))
sns.barplot(y='track_name', x='streams', data=top_tracks, palette='Blues_d')
plt.title('Top 10 Tracks por Streams', fontsize=16)
plt.xlabel('Streams', fontsize=12)
plt.ylabel('Track Name', fontsize=12)
plt.tight_layout()

# 4. Artist(s)_name por Streams (Top 10 artistas por Streams)
top_artists = df[['artist(s)_name', 'streams']].sort_values(by='streams', ascending=False).head(10)
plt.figure(figsize=(10, 6))
sns.barplot(y='artist(s)_name', x='streams', data=top_artists, palette='Blues_d')
plt.title('Top 10 Artists por Streams', fontsize=16)
plt.xlabel('Streams', fontsize=12)
plt.ylabel('Artist(s)_name', fontsize=12)
plt.tight_layout()

# Exibindo todos os gráficos
plt.show()
```

# Insights e Recomendações para o Negócio

Aqui estão alguns **insights** e **recomendações** que podem ser úteis para a análise de dados e para fornecer valor tanto em termos de **lucratividade** quanto em **qualidade de vida para os clientes**:

### 1. **Diversidade e Popularidade das Músicas**
- **Insight:** A variável **`artist_count`** apresenta uma média de **1,56**, o que significa que, em média, as músicas são lançadas por 1 ou 2 artistas (ou seja, colaborações são comuns, mas a maioria das músicas é de um artista solo). Além disso, o valor máximo é **8** artistas, indicando algumas colaborações com múltiplos artistas.
  
  - **Recomendação para Negócio:** Investir em parcerias e colaborações entre artistas pode ser uma estratégia lucrativa, pois os lançamentos colaborativos podem atrair diferentes públicos e aumentar os streams. Além disso, promover mais músicas de artistas solos e permitir maior visibilidade para novos artistas também pode ser uma boa estratégia.

---

### 2. **Tendências de Lançamentos**
- **Insight:** O ano médio de lançamento das músicas é **2018** e a média de lançamentos ocorre nos meses **junho (6)** e **novembro (11)**, conforme o valor da variável **`released_month`**. A distribuição de datas de lançamento é bastante ampla, com lançamentos ocorrendo desde 1930 até 2023.
  
  - **Recomendação para Negócio:** Estratégias de **marketing e lançamento** podem se beneficiar do foco nas tendências sazonais de lançamentos. O mês de junho e novembro parece ser um ponto chave, então, campanhas de lançamento ou promoções poderiam ser planejadas nessas datas para maximizar a audiência e o engajamento. Além disso, aproveitar a nostalgia de músicas mais antigas pode ser uma forma de atrair um público mais amplo.

---

### 3. **Streams e Popularidade**
- **Insight:** A média de **streams** é **513.597.900** e o valor máximo é **3.703.895.000**. Esses valores indicam uma grande variação na popularidade das músicas, com algumas músicas alcançando números extraordinários de streams, enquanto outras não têm tanto sucesso.
  
  - **Recomendação para Negócio:** Investir em estratégias de **promoção para músicas menos populares** pode ajudar a equilibrar a distribuição de streams. É importante identificar padrões que levam uma música a ser mais popular e replicá-los para novos lançamentos. Além disso, a **curadoria de playlists** para músicas com menor número de streams pode aumentar sua exposição e, consequentemente, seus números.

---

### 4. **BPM (Batidas Por Minuto) e Características das Músicas**
- **Insight:** O BPM médio das músicas é **122**, o que é um valor relativamente alto e sugere que a maior parte das músicas pode ter um ritmo acelerado (tipicamente mais dançante). As músicas têm valores médios de **danceability (%)** em **66,97%** e **energy (%)** em **64,28%**, o que reforça a ideia de que a maioria das músicas possui uma energia média a alta, ideal para dançar.
  
  - **Recomendação para Negócio:** Criar playlists e campanhas de marketing com foco em músicas **dançantes e energéticas** pode ser uma boa estratégia, especialmente se alinhada com eventos e datas comemorativas que incentivem atividades mais dinâmicas, como festas, eventos esportivos ou promoções voltadas para o público jovem.

---

### 5. **Diversidade Sonora e Qualidade das Músicas**
- **Insight:** O valor médio de **acousticness** é **27,06%**, indicando que a maioria das músicas não são acústicas, sendo mais voltadas para produções mais eletrônicas ou com uso de instrumentos elétricos. A **valence (%)** média de **51,43%** sugere um tom de humor equilibrado nas músicas, sem ser excessivamente positivo nem negativo.
  
  - **Recomendação para Negócio:** Aproveitar o **equilíbrio emocional** das músicas (como indicado pela **valence** média) para criar playlists que atendam a diferentes estados de espírito. As playlists de **ambientes relaxantes** ou **acústicas** podem atrair um público que busca algo mais suave, ao passo que músicas com uma maior **energia** e **ritmo** podem atrair o público que busca mais dinamismo.

---

### 6. **Engajamento nas Plataformas**
- **Insight:** A maior parte das músicas está sendo tocada em **Spotify (em playlists e charts)**, seguido por **Apple Music** e **Deezer**, com destaque para o **Shazam** em termos de descobertas de músicas. A média de músicas nas **playlists do Spotify** é **5200**, indicando uma boa presença, mas com grande variação (de 31 até 52.898).
  
  - **Recomendação para Negócio:** **Expandir parcerias com plataformas de streaming** e aumentar a presença em **playlists populares** do Spotify e outras plataformas. Investir em campanhas para **aumentar a presença em playlists** e melhorar a visibilidade nas **charts** pode ajudar a maximizar a exposição das músicas e aumentar os streams. Além disso, se possível, explorar o uso do **Shazam** pode ser uma boa forma de impulsionar a descoberta das músicas.

---

### 7. **Outliers e Oportunidades de Crescimento**
- **Insight:** O **máximo de streams** e o **número de músicas em playlists** apresentam grandes outliers (valores extremamente altos comparados à média), o que pode indicar que algumas músicas estão dominando o mercado, enquanto outras estão sendo negligenciadas.
  
  - **Recomendação para Negócio:** Identificar os **outliers** e tentar entender o que fez essas músicas tão populares. Pode ser interessante analisar os padrões de sucesso (bpm, energia, tipo de artista, etc.) e replicá-los para outras músicas. Além disso, explorar a **diversidade de lançamentos** (por exemplo, músicas acústicas, mais suaves, ou de outros gêneros) pode ajudar a alcançar uma audiência mais ampla e diversificada, sem depender apenas dos grandes sucessos.

---

### Conclusão

**Para o negócio:**
- Focar em **parcerias de artistas**, especialmente colaborativos, pode ajudar a ampliar a audiência.
- Aproveitar as tendências de **lançamentos sazonais**, especialmente em meses como junho e novembro, pode aumentar a visibilidade e engajamento.
- **Investir em campanhas de marketing** direcionadas para músicas com menos streams, mas com potencial de crescimento.
- **Criar playlists dinâmicas** que atendam tanto ao público que gosta de músicas enérgicas quanto ao público que prefere músicas mais calmas e acústicas.

Essas ações podem contribuir para o **aumento dos streams**, melhor **experiência do usuário** e, consequentemente, **maior lucratividade** para a empresa.
